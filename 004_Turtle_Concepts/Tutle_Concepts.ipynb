```python
import turtle
```

```python

i = turtle.Turtle() 
i.pen(pencolor="red", fillcolor="yellow")
i.shapesize(10, 5, 2)   # To change the turtle size
i.stamp()
i.begin_fill()
i.circle(90)
i.end_fill()

turtle.mainloop() # to keep the screen awake
```
