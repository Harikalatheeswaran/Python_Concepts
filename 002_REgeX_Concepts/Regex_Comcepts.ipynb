### <center> Important Notations
- " ^ " : is used to find/match the begining of a line. eg: ^From:
- " \$ " : is used to find/match the end of a line. eg: From:$
- " \\S " : is used to find/match a non-white space
- " \\s " : is used to find/match a white space
- Wild card characters: ".", matches any character
- The dot charcter matches any character.
- "\[aeiou\]" : matches a single character in a single set.
- "\[^XYZ\]" : matches a single character not in a single set.
- " .* " : matches any number of characters
- " * "  : implies any number of times
- " + "  : implies match one or more characters present atleast once or more times.
- " * "  : implies match one or more characters present zero or more times.
- _Eg : "^X.*:", begining with X and .\* means match any number of characters._

```python
import re
```

```python
# re.search() and re.findall() are most commonly used.
line = "From: etdhari@gmail.com"
print(re.search("From:", line))
print(re.search("^From:", line)) # "^": is used to find the begining of a line
if re.search('^From:', line):
    print(line)
if re.search('.com$', line):
    print(line)
if re.search('^F..m', line): # .. implies match any (2) character between F and m
    print(line)
```

```python
s = "A message from csev@umich.edu to cwen@iupui.edu about the meeting @2pm"
l1 = re.findall("\S+@\S", s)
l2 = re.findall("\S+@\S+", s)
l3 = re.findall("\S*@\S+", s)
print("l1: ", l1)
print("l2: ",l2)
print("l3: ", l3)
```

```python
x = "My 2 favourite numbers are 19 and 42,  1.23"
y1 = re.findall('[0-9]', x)
y2 = re.findall('[0-9]*', x)
y3 = re.findall('[0-9].', x)
y4 = re.findall('[0-9]+', x)
y5 = re.findall('[0-9.]+', x) # <--- period to obtain floating point numbers
y6 = re.findall('[0-9][0-9]+', x) # <-- To find the two digit numbers

print("y1:", y1)
print("y2:", y2)
print("y3:", y3)
print("y4:", y4)
print("y5:", y5)
print("y6:", y6)
```

```python
x2 = "From: Using : hari: 662000"
greedy = re.findall('^F.+:' , x2)
print("Greedy: ", greedy)
# since we have the ".+" the first ":" is also a character, it finds till last colon.
# It's called a "greedy match"
# If we want only till "From:", the we do non greedy search by using "?" symbol before a + symbol:
non_greedy = re.findall('^F.+?:' , x2)
print("Non greedy: ", non_greedy)
```

```python
# "()" are used to start and stop extaraction.
x3 = 'hello myman@hari w@asup'
print(re.findall("\S+@\S+", x3)) # '\S+' is used to match atleast one non white space character.
print(re.findall("[^ ]", x3))
print(re.findall("\s.+\s", x3))
# to obtain th e host name
print(re.findall("@([a-z.]+)", x3))
```
