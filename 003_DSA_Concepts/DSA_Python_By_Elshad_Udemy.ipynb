```python
# Eg 1
def f1():
    f2()
    print("f1")

def f2():
    f3()
    print("f2")

def f3():
    f4()
    print("f3")

def f4():
    print("f4")

f1()
```

```python
#Eg 2
def recursiveMethod(n):
    if n<1:
        print("n<1")
    else:
        recursiveMethod(n-1)
        print(n)
recursiveMethod(6)
```

```python
# Recursion  to find power of 2
def pow2(n):
    if n == 0:
        return 1
    else:
        p = pow2(n-1)
        return p*2

pow2(5)
```

```python
# Recursion to find the factorial

def fact(n):
    assert n>0 and type(n)==int, "Enter positive integers only"
    if n in [0,1] :
        return 1
    else:
        return n*fact(n-1)
print(fact(-3))
```

```python
# finding the nth fibbonacci number
def fib(n):
    assert (n>=0 and type(n)==int), "Enter positive integer only"
    if n in (0,1):
        return n
    else:
        return fib(n-1)+fib(n-2)
print(fib(5))
```

```python
# Sum of posiive digits
def sum_of_dig(n):
    assert (n>=0 and type(n)==int), "Enter positive integer only"
    if n in (1,2,3,4,5,6,7,8,9):
        return n
    else:
        return n%10 + sum_of_dig(n//10)
print(sum_of_dig(12345))
```

```python
# GCD of two numbers
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a%b)
print(gcd(18, 48))
```

```python
# decimal to binary
def dtb(n):
    if n == 0:
        return n
    else:
        x = n%2 +10*dtb(n//2)
        return x
        
print(dtb(9))
```
