```python
# GPT
from pynput import keyboard
from pynput.keyboard import Controller, Key


# Initialize the Controller
kb = Controller()

# Define your list
my_list = ['Hello', 'World', 'Python', 'Programming']

def form_filler(data : list):
    """Function to perform form filling. Note: press Esc to terminate the program."""
    def on_press(key):
        # Check if the key pressed is a left arrow key
        if key == Key.right:
            try:
                if my_list != None:
                    # Enter the first item from the list
                    kb.type(data[0])
                    # Remove the first item from the list
                    my_list.pop(0)
                else:
                    return False
            except:
                print("End of list")
        if key == Key.delete or key == Key.esc:
            return False

    # Collect events until released
    with keyboard.Listener(on_press=on_press) as listener:
        listener.join()

form_filler(my_list)
```

Here's a basic example of how to use pynput to print out the keys pressed:

```python
from pynput import keyboard

def on_press(key):
    try:
        # Print the key that was pressed
        print(f'Key {key.char} pressed')
    except AttributeError:
        # Handle special keys
        print(f'Special key {key} pressed')

def on_release(key):
    # Stop listener with ESC key
    if key == keyboard.Key.esc:
        return False

# Start listening to the keyboard
with keyboard.Listener(on_press=on_press, on_release=on_release) as listener:
    listener.join()
```

Pressing the trigger key 'F9' to start a certain key board & mouse simulation & pressing 'ESC' key to stop those actions.

```python
from pynput import keyboard, mouse
import time

# Define the key to trigger the actions
trigger_key = keyboard.Key.f9

# Initialize the mouse controller
mouse_controller = mouse.Controller()

# Define the actions to perform
def perform_actions():
    # Example: move the mouse to (100, 100) and click
    mouse_controller.position = (100, 100)
    mouse_controller.click(mouse.Button.left)
    
    # Example: press and release the 'a' key
    keyboard_controller = keyboard.Controller()
    keyboard_controller.press('a')
    time.sleep(0.1)  # Small delay
    keyboard_controller.release('a')

# Define the callback function for key press events
def on_press(key):
    if key == trigger_key:
        perform_actions()

# Define the callback function for key release events
def on_release(key):
    if key == keyboard.Key.esc:
        # Stop listener
        return False

# Start listening to keyboard events
with keyboard.Listener(on_press=on_press, on_release=on_release) as listener:
    listener.join()
```

Okay, help me create python program which will first record the mouse & keyboard actions performed after pressing say a trigger key called trigger_key_1 press esc button to stop recording.  then press another trigger key called trigger_key_2 to mimic the same moues & key stroke actions which was earlier performed after pressing triiger_key_1 

```python
from pynput import keyboard, mouse
from pynput.mouse import Button, Controller as MouseController
from pynput.keyboard import Key, Controller as KeyboardController
import time

# Initialize global variables
recorded_events = []
is_recording = False

# Initialize controllers
mouse_controller = MouseController()
keyboard_controller = KeyboardController()

def on_press(key):
    global is_recording, recorded_events

    if key == Key.esc:
        # Stop recording
        is_recording = False
        print("Recording stopped.")
        return False  # This stops the listener

    if is_recording:
        try:
            recorded_events.append(('key_press', key.char, time.time()))
        except AttributeError:
            recorded_events.append(('key_press', str(key), time.time()))

    if key == Key.f9:  # Trigger key for recording
        is_recording = True
        recorded_events = []
        print("Recording started.")

    if key == Key.f10:  # Trigger key for playback
        print("Playback started.")
        play_recorded_events()

def on_click(x, y, button, pressed):
    global is_recording, recorded_events

    if is_recording:
        action = 'pressed' if pressed else 'released'
        recorded_events.append(('mouse_click', x, y, button, action, time.time()))

def on_move(x, y):
    global is_recording, recorded_events

    if is_recording:
        recorded_events.append(('mouse_move', x, y, time.time()))

def play_recorded_events():
    if not recorded_events:
        print("No events to replay.")
        return

    start_time = recorded_events[0][-1]

    for event in recorded_events:
        event_type, *data, timestamp = event
        elapsed_time = timestamp - start_time

        time.sleep(elapsed_time)
        
        if event_type == 'key_press':
            key = data[0]
            if key.startswith('Key.'):
                key = getattr(Key, key[4:])
            keyboard_controller.press(key)
            keyboard_controller.release(key)
        elif event_type == 'mouse_click':
            x, y, button, action = data
            if action == 'pressed':
                mouse_controller.position = (x, y)
                mouse_controller.press(button)
            else:
                mouse_controller.position = (x, y)
                mouse_controller.release(button)
        elif event_type == 'mouse_move':
            x, y = data
            mouse_controller.position = (x, y)

# Set up listeners
keyboard_listener = keyboard.Listener(on_press=on_press)
mouse_listener = mouse.Listener(on_click=on_click, on_move=on_move)

# Start listeners
keyboard_listener.start()
mouse_listener.start()

keyboard_listener.join()
mouse_listener.join()
```

Explanation
Recording Event:

on_press(key): Records key presses when trigger_key_1 (F9) is pressed. Stops recording when ESC is pressed. <br>
on_click(x, y, button, pressed): Records mouse clicks with positions and button states.<br>
on_move(x, y): Records mouse movement with positions.<br>
Playback Event:<br>

play_recorded_event(): Replays the recorded event, simulating the mouse movement and key presses according to the recorded timestamps.<br>
Trigger Keys:<br>

F9: Start recording.<br>
F10: Start playback.<br>

# <center> **Alt-Codes**

```python
from pynput.keyboard import Controller, Key
import time

# Initialize keyboard controller
keyboard = Controller()

def alt_type_sequence(input_str: str, delay: float = 0.0001):
    """
    Simulates tapping all digits in the input string while holding Alt.
    Optimized for Jupyter Notebook to avoid tab switching.
    Args:
        input_str: String of digits to tap (e.g., "123").
        delay: Delay between key actions to ensure reliability (default 0.01s).
    """
    print("Waiting 3 seconds to place the cursor...")
    time.sleep(3)
    
    # Validate input is numeric
    if not input_str.isdigit():
        print("Error: Input must be numeric!")
        return None
    
    try:
        # Hold Alt key during digit taps
        with keyboard.pressed(Key.alt):
            # Tap each digit in the input string
            for digit in input_str:
                keyboard.tap(digit)
                print(f"Tapped {digit}")
                time.sleep(delay)  # Small delay for reliability
        
        print("Sequence completed!")
        
    except ValueError as e:
        print(f"Error: Invalid key in sequence - {e}")
        return None

```

```python
# Example usage
if __name__ == "__main__":
    alt_type_sequence("123")
```
