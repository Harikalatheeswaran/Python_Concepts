````python

# Python program showinguse of format() method
 
# using format() method
print('I love {} for "{}!"'.format('Geeks', 'Geeks'))
 
# using format() method and referring a position of the object
print('{0} and {1}'.format('Geeks', 'Portal'))
 
print('{1} and {0}'.format('Geeks', 'Portal'))
 
# the above formatting can also be done by using f-Strings
# Although, this features work only with python 3.6 or above.
 
print(f"I love {'Geeks'} for \"{'Geeks'}!\"")
 
# using format() method and referring a position of the object
print(f"{'Geeks'} and {'Portal'}")

# using a variable:
n = 69
print("what do you mean by {}".format(n))
print(f"what do you mean by {n}")
````

### <u> Regular Expressions </u>

#### Basic Functions
* <u>re.match(pattern, string) :</u><br>
Checks for a match only at the beginning of the string.
Returns a match object if the pattern matches; otherwise, it returns None.
```
import re
result = re.match(r'hello', 'hello world')
print(result.group())  # Output: hello
```

* <u>re.search(pattern, string) :</u><br>
Searches the entire string for the first occurrence of the pattern.
Returns a match object if the pattern is found; otherwise, it returns None.

```
import re
result = re.search(r'world', 'hello world')
print(result.group())  # Output: world
```
* <u>re.findall(pattern, string) :</u><br>
Returns a list of all matches of the pattern in the string.
```
import re
results = re.findall(r'\d+', 'There are 12 apples and 34 oranges.')
print(results)  # Output: ['12', '34']
```
* <u>re.finditer(pattern, string) :</u><br>
Returns an iterator yielding match objects for all matches in the string.
```
import re
for match in re.finditer(r'\d+', 'There are 12 apples and 34 oranges.'):
    print(match.group())  # Output: 12, then 34
```

* <u>re.sub(pattern, repl, string) :</u><br>
Replaces occurrences of the pattern in the string with repl.


```
import re
result = re.sub(r'\d+', 'number', 'There are 12 apples and 34 oranges.')
print(result)  # Output: There are number apples and number oranges.
```
```
Regular Expression Syntax
Literal Characters: Match themselves, e.g., a matches 'a'.

Metacharacters: Special characters with special meanings, such as:

. : Matches any character except a newline.
^ : Matches the start of the string.
$ : Matches the end of the string.
* : Matches 0 or more repetitions.
+ : Matches 1 or more repetitions.
? : Matches 0 or 1 repetition.
[] : Matches any one of the characters inside the brackets.
| : Acts as a logical OR.
()  : Groups patterns.
\d : Matches any digit (equivalent to [0-9]).
\w : Matches any word character (equivalent to [a-zA-Z0-9_]).
\s : Matches any whitespace character (space, tab, newline).
```


Example Usage
Heres an example that uses a few of these functions and concepts:
```
import re

text = "The price of the book is $25 and the pen is $2."

# Find all dollar amounts
amounts = re.findall(r'\$\d+', text)
print(amounts)  # Output: ['$25', '$2']

# Replace dollar amounts with 'PRICE'
modified_text = re.sub(r'\$\d+', 'PRICE', text)
print(modified_text)  # Output: The price of the book is PRICE and the pen is PRICE.
```

Word Boundary (\b)
```
Definition: \b asserts a position where one side is a word character (\w), and the other side is a non-word character (\W) or the start/end of the string. It does not consume any characters in the string but rather checks the position.

Word Characters: \w matches any alphanumeric character (letters and digits) and underscores (_).

Non-Word Characters: \W matches any character that is not a word character, such as spaces, punctuation, etc.

```

```
Explanation:
loading_animation function: Displays the loading animation using a cycling sequence of characters (|, /, -, \).
main_function: Simulates the long-running operation (you can replace time.sleep(10) with your actual task).
Threading: The loading animation runs in a separate thread so that it continues while the main function is being executed.
```

````python
import time
import sys
import itertools
import threading

# Function to display loading animation
def loading_animation():
    animation = itertools.cycle(['|', '/', '-', '\\'])  # Characters to cycle through
    while not stop_loading:  # Loop until stop signal is given
        sys.stdout.write(f"\rLoading... {next(animation)}")
        sys.stdout.flush()
        time.sleep(0.1)  # Adjust sleep to control speed of animation

# Main function that performs long-running operation
def main_function():
    # Simulate long-running task (replace this with your actual function)
    time.sleep(10)  # Example delay for the long-running process

# Variable to stop the animation
stop_loading = False

# Start the loading animation in a separate thread
loading_thread = threading.Thread(target=loading_animation)
loading_thread.start()

# Run the long-running function
main_function()

# Stop the animation after the function finishes
stop_loading = True
loading_thread.join()

# Print completion message
print("\nProcess complete!")
````

```
Dot Loading Animation: The loading_animation function now prints Loading... followed by an increasing number of dots (up to 3), then resets and repeats.

This version adds a subtle but visually pleasing animation while your process is running!
```

````python

import time
import sys
import threading

# Function to display dot loading animation
def loading_animation():
    dots = 0
    while not stop_loading:
        sys.stdout.write(f"\rLoading{'.' * dots}")
        sys.stdout.flush()
        dots = (dots + 3) % 4 # Cycle through 0, 1, 2, 3 dots
        time.sleep(0.69)  # Adjust sleep time for animation speed

# Main function that performs long-running operation
def main_function():
    # Simulate long-running task (replace this with your actual function)
    time.sleep(5)  # Example delay for the long-running process

# Variable to stop the animation
stop_loading = False

# Start the loading animation in a separate thread
loading_thread = threading.Thread(target=loading_animation)
loading_thread.start()

# Run the long-running function
main_function()

# Stop the animation after the function finishes
stop_loading = True
loading_thread.join()

# Print completion message
print("\nProcess complete!")
````

```
Explanation:
Progress(): This is a rich class that allows you to display a progress bar in the terminal. The Progress object manages one or more tasks (e.g., loading bars) that you can update dynamically.

add_task(): This creates a new task with a label ("[cyan]Processing...") and specifies a total number of units of work (total=100). You can adjust the total based on your actual task's complexity.

update(): This method updates the progress bar by incrementing the progress (advance=1), which means it's advancing one step each time. You can adjust how much work is done in each step by changing the advance parameter.

sleep(0.1): This simulates time being spent on each step of the process. In your actual function, replace this with real data processing steps.
```

````python
from time import sleep
from rich.progress import Progress

# Main function that performs a long-running task
def main_function():
    total_work = 100  # Simulated total work units
    for i in range(total_work):
        sleep(0.1)  # Simulate work being done
        yield i  # Simulate work progress

# Display progress bar using rich
def display_progress_bar():
    with Progress() as progress:
        task = progress.add_task("[cyan]Processing...", total=100)
        
        # Simulate the long-running function and update progress
        for step in main_function():
            progress.update(task, advance=1)

# Run the progress bar while the main function executes
display_progress_bar()

print("Process complete!")
````

### Printing with a delay 

````python
import time

def print_with_delay(text, t):
    for char in text:
        print(char, end='', flush=True)
        time.sleep(t)

# Example usage
text = "Hello, World!"
delay = 0.25  # 0.25 seconds delay between each letter
print_with_delay(text, delay)
````

# **RICH Module Hacks!**

````python
from tqdm import tqdm
from time import sleep


for i in tqdm(range(100),
    desc="Downloading test files",
    unit="file",
    bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt} [{elapsed}<{remaining}]',
    mininterval=0.5,
):
    # do something
    sleep(0.1)
    pass
````

````python
from rich.progress import track
from time import sleep

for link in track(range(99), description="Joshi...", transient=False):
    sleep(0.1)
````

 Parameter Details<br>
* Parameter	Description<br>
* sequence	The iterable to loop over (e.g., range(100) or a list of links).<br>
* total	Total number of items (optional if sequence is a list or has __len__).<br>
* description	Text shown to the left of the progress bar.<br>
* update_period	How often the display updates (in seconds). Default is 0.1s.<br>
* transient	If True, progress bar disappears after completion.<br>
