##### <u>*__1. Creating a PDF__*</u>

```python
from reportlab.pdfgen import canvas

def create_pdf():
    # create canvas object
    pdf_canvas = canvas.Canvas("report.pdf")

    # Set the title of pDF
    pdf_canvas.setTitle("Report")

    # Draw a stirng on the pDF
    pdf_canvas.drawString(0, 720, "a"*87)
    pdf_canvas.drawString(90, 730, "Heeeeello World"*9)

    #Save PDF
    pdf_canvas.save()

create_pdf()
```
