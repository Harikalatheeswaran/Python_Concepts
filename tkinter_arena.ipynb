<u>**_Label Widget:_** <br></u><br>
Tkinter Label is a widget that is used to implement display boxes where you can place text or images. The text displayed by this widget can be changed by the developer at any time you want. It is also used to perform tasks such as to underline the part of the text and span the text across multiple lines. It is important to note that a label can use only one font at a time to display text. To use a label, you just have to specify what to display in it (this can be text, a bitmap, or an image).<br>
**Syntax :**
> *w = Label ( master (root), option, ‚Ä¶ )*

* <u>master : <br></u> This represents the parent window
* <u>options : <br></u> Below is the list of most commonly used options for this widget. These options can be used as key-value pairs separated by commas : <br></u>
<br>
**Various Options are :** <br>

* <u>anchor : <br></u> This options is used to control the positioning of the text if the widget has more space than required for the text. The default is an<u>chor=CENTER, which centers the text in the available space.
* bg : <br></u>This option is used to set the normal background clior displayed behind the label and indicator.
* <u>height : <br></u>This option is used to set the vertical dimension of the new frame.
* <u>width : <br></u>Width of the label in characters (not pixels!). If this option is not set, the label will be sized to fit its contents.
* bd : <br></u>This option is used to set the size of the border around the indicator. Default bd value is set on 2 pixels.
* <u>font : <br></u>If you are displaying text in the label (with the text or textvariable option), the font option is used to specify in what font that text in the label will be displayed.
cu<u>rsor : <br></u>It is used to specify what cursor to show when the mouse is moved over the label. The default is to use the standard cursor.
te<u>xtvariable : <br></u> As the name suggests it is associated with a Tkinter variable (usually a StringVar) with the label. If the variable is changed, the label text is updated.
* <u>bitmap : <br></u>It is used to set the bitmap to the graphical object specified so that, the label can represent the graphics instead of text.
* fg : <br></u>The label clior, used for text and bitmap labels. The default is system specific. If you are displaying a bitmap, this is the clior that will appear at the position of the 1-bits in the bitmap.
* <u>image : <br></u> This option is used to display a static image in the label widget.
* <u>padx : <br></u>This option is used to add extra spaces between left and right of the text within the label.The default value for this option is 1.
* <u>pady : <br></u>This option is used to add extra spaces between top and bottom of the text within the label.The default value for this option is 1.
* <u>justify : <br></u>This option is used to define how to align multiple lines of text. Use LEFT, RIGHT, or CENTER as its values. Note that to position the text inside the widget, use the anchor option. Default value for justify is CENTER.
* <u>relief : <br></u> This option is used to specify appearance of a decorative border around the label. The default value for this option is FLAT.
* <u>underline : <br></u>This
* <u>wraplength : <br></u>Instead of having only one line as the label text it can be broken into to the number of lines where each line has the number of characters specified to this option.

<u>**_Combo Box:_** <br></u><br>
It's a combination of Listbox and an entry field. It is one of the Tkinter widgets where it contains a down arrow to select from a list of options. It helps the users to select according to the list of options displayed. When the user clicks on the drop-down arrow on the entry field, a pop up of the scrolled Listbox is displayed down the entry field. The selected option will be displayed in the entry field only when an option from the Listbox is selected.<br>
**Syntax :**
> *combobox = ttk.Combobox(master, option=value, ...)*

```python
# python program demonstrating
# Combobox widget using tkinter


import tkinter as tk
from tkinter import ttk

# Creating tkinter window
window = tk.Tk()
window.title('Combobox')
window.geometry('500x250')

# label text for title (optional)
ttk.Label(window, text = "GFG Combobox Widget",
		background = 'green', foreground ="white",
		font = ("Times New Roman", 15)).grid(row = 0, column = 1)

# label
ttk.Label(window, text = "Select the Month :",
		font = ("Georgia", 10)).grid(column = 0,
		row = 5, padx = 10, pady = 25)

# Combobox creation
n = tk.StringVar()
monthchoosen = ttk.Combobox(window, width = 27, textvariable = n)

# Adding combobox drop down list
monthchoosen['values'] = [' January',
						' February',
						' March',
						' April',
						' May',
						' June',
						' July',
						' August',
						' September',
						' October',
						' November',
						' December']

monthchoosen.grid(column = 1, row = 5)
close_button = ttk.Button(window,  text = "Close", command=window.destroy).place(x = 90, y = 162)
print("\r", monthchoosen.current(), str(n.get()))
print(dir(n))
window.mainloop()
```

<u>**_maxsize():_** <br></u><br>
This method is used to set the maximum size of the root window (maximum size a window can be expanded). User will still be able to shrink the size of the window to the minimum possible.<br>
**Syntax :**
> *master.maxsize(height, width)* <br>



```python
# importing only those functions
# which are needed
from tkinter import *
from tkinter.ttk import *

# creating tkinter window
root = Tk()

# Fixing the size of the root window.
# No one can now expand the size of the
# root window than the specified one.
root.maxsize(600, 600)

# Adding widgets to the root window
Label(root, text = 'GeeksforGeeks',
	font =('Verdana', 15)).pack(side = TOP, pady = 10)

Button(root, text = 'Close', command=root.destroy).pack(side = TOP)

mainloop()
```

<u>**_Entry Widget:_** <br></u><br>
The Entry Widget is a Tkinter Widget used to Enter or display a single line of text. <br>
**Syntax :**
> *entry = tk.Entry(parent, options)*

<u>**_Parameters:_** <br></u><br>
 

1) Parent: The Parent window or frame in which the widget to display. eg: root
2) Options: The various options provided by the entry widget are: 
 

* **_bg :_** The normal background color displayed behind the label and indicator. <br>
* **_bd :_** The size of the border around the indicator. Default is 2 pixels. <br>
* **_font :_** The font used for the text. <br>
* **_fg :_** The color used to render the text. <br>
* **_justify :_** If the text contains multiple lines, this option controls how the text is justified: CENTER, LEFT, or RIGHT. <br>
* **_relief :_** With the default value, relief=FLAT. You may set this option to any of the other styles like : SUNKEN, RIGID, RAISED, GROOVE <br>
* **_show :_** Normally, the characters that the user types appear in the entry. To make a .password. entry that echoes each character as an asterisk, set show=‚Äù*‚Äù. <br>
* **_textvariable :_** In order to be able to retrieve the current text from your entry widget, you must set this option to an instance of the StringVar class.<br>

**Methods: The various methods provided by the entry widget are:**

**_get() :_** Returns the entry‚Äôs current text as a string. <br>
**_delete() :_** Deletes characters from the widget <br>
**_insert ( index, ‚Äòname‚Äô) :_** Inserts string ‚Äòname‚Äô before the character at the given index. <br>

```python
# Program to make a simple
# login screen


import tkinter as tk

root=tk.Tk()

# setting the windows size
root.geometry("600x400")

# declaring string variable for storing name and password
name_var=tk.StringVar()
passw_var=tk.StringVar()
company=tk.StringVar()
company.set("Continental")

# defining a function that will get the name and password and print them on the screen
def submit():
	name=name_var.get()
	password=passw_var.get()
	
	
	print("The name is : " + name)
	print("The password is : " + password)
	print(company)
	
	name_var.set("")
	passw_var.set("")
    
# creating a label forname using widget Label
company_label = tk.Label(root, text = 'Company',state='disabled', font=('Georgia',10, 'bold'))
# creating a Read only entry for company using widget Entry
company_entry = tk.Entry(root, textvariable = company, font=('Georgia',10,'normal')) # Read only Entry widget

# creating a label forname using widget Label
name_label = tk.Label(root, text = 'Username', font=('Georgia',10, 'bold'))
# creating a entry for input name using widget Entry
name_entry = tk.Entry(root, textvariable = name_var, font=('Georgia',10,'normal'))

# creating a label for password
passw_label = tk.Label(root, text = 'Password', font = ('Georgia',10,'bold'))
# creating a entry for password
passw_entry=tk.Entry(root, textvariable = passw_var, font = ('Georgia',10,'normal'), show = '*')

# creating a button using the widget Button that will call the submit function
sub_btn=tk.Button(root,text = 'Submit', command = submit)

# placing the label and entry in the required position using grid method
name_label.grid(row=0,column=0)
name_entry.grid(row=0,column=1)
passw_label.grid(row=1,column=0)
passw_entry.grid(row=1,column=1)
company_label.grid(row=2, column=0)
company_entry.grid(row=2, column=1)
sub_btn.grid(row=3,column=1)
#
# performing an infinite loop for the window to display
root.mainloop()
```

<u>**ChekButton Widget:_** <br></u><br>
The Checkbutton widget is a standard Tkinter widget that is used to implement on/off selections. Checkbuttons can contain text or images. When the button is pressed, Tkinter calls that function or method.

Syntax:
The syntax to use the checkbutton is given below.. <br>
**Syntax :**
> *check_button = tk.Checkbutton(parent, options)*

Methods used in this widget is as follows: <br>
* **deselect()**: This method is called to turn off the checkbutton. <br>
* **flash()**: The checkbutton is flashed between the active and normal colors.<br>
* **invoke()**: This method will invoke the method associated with the checkbutton.<br>
* **select()**: This method is called to turn on the checkbutton.<br>
* **toggle()**: This method is used to toggle between the different Checkbuttons.<br>

Check this link for more info on Checkbutton widget: [Link](https://www.geeksforgeeks.org/python-tkinter-checkbutton-widget/)

```python
from tkinter import *

root = Tk()
root.geometry("300x200")

w = Label(root, text ='GeeksForGeeks', font = "50")
w.pack()

Checkbutton1 = IntVar()
Checkbutton2 = IntVar()

Button1 = Checkbutton(root,
					variable = Checkbutton1,
					onvalue = 1,
					offvalue = 0,
					height = 2,
					width = 10)

Button2 = Checkbutton(root, text = "Student",
					variable = Checkbutton2,
					onvalue = 1,
					offvalue = 0,
					height = 2,
					width = 10)

	
Button1.pack()
Button2.pack()

mainloop()

```

```python
import tkinter as tk
from tkinter import Toplevel
from tkcalendar import Calendar

def open_calendar():
    # Create a new top-level window for the calendar
    calendar_window = Toplevel(root)
    calendar_window.title("Select Date")

    # Create and pack the calendar widget in the top-level window
    cal = Calendar(calendar_window, selectmode='day', year=2024, month=8, day=21)
    cal.pack(pady=20)

    # Function to fetch and display the selected date
    def select_date():
        selected_date = cal.get_date()
        date_label.config(text=f"Selected Date: {selected_date}")
        calendar_window.destroy()  # Close the calendar window

    # Create a button to confirm the date selection
    select_button = tk.Button(calendar_window, text="Select Date", command=select_date)
    select_button.pack(pady=10)

# Create the main window
root = tk.Tk()
root.title("Date Picker")

# Create a button to open the calendar
open_calendar_button = tk.Button(root, text="Open Calendar", command=open_calendar)
open_calendar_button.pack(pady=20)

# Create a label to display the selected date
date_label = tk.Label(root, text="Selected Date:")
date_label.pack(pady=10)

# Run the application
root.mainloop()
```

```python
import tkinter as tk
from tkinter import Toplevel
from tkcalendar import Calendar

def open_calendar(event):
    # Create a new top-level window for the calendar
    calendar_window = Toplevel(root)
    calendar_window.title("Select Date")

    # Create and pack the calendar widget in the top-level window
    cal = Calendar(calendar_window, selectmode='day', year=2024, month=8, day=21)
    cal.pack(pady=20)

    # Function to fetch and display the selected date
    def select_date():
        selected_date = cal.get_date()
        date_entry.delete(0, tk.END)  # Clear current text in the entry
        date_entry.insert(0, selected_date)  # Insert selected date
        calendar_window.destroy()  # Close the calendar window

    # Create a button to confirm the date selection
    select_button = tk.Button(calendar_window, text="Select Date", command=select_date)
    select_button.pack(pady=10)

# Create the main window
root = tk.Tk()
root.title("Date Picker")

# Create an Entry widget to trigger the calendar
date_entry = tk.Entry(root, width=20)
date_entry.bind("<Button-1>", open_calendar)  # Bind the left mouse button click event to open the calendar
date_entry.pack(pady=20)

# Run the application
root.mainloop()
```
